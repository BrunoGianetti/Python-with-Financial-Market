**Iniciando as bibliotecas e puxando os dados de Mercado Financeiro**

```python
import pandas_datareader as pdr
import datetime
import numpy as np

vale3 = pdr.get_data_yahoo('VALE3.SA', start=datetime.datetime(2016,10,1), end=datetime.datetime(2019,1,1))

# Plotando toda a biblioteca

vale3
```

```python
# Retornando os índices do DataFrame

print(vale3.index)

# Como trabalharemos com séries temporais, os índices serão quase sempre datas.
```

```python
# Retorando as colunas do DataFrame

print(vale3.columns)
```

```python
# Para filtrar os períodos que queremos analisar através do índice do DataFrame. Um cutoff.

# Exemplo: retornando os 10 primeiros registros a partir de 01/01/2018

vale3['2018-01-01':].head(10)
```

```python
# Para calcular a variação percentual intradiária do ativo

# cálculo: preço de fechamento / preço de abertura

# Com o pandas dividimos as colunas de forma direta.

vale3.Close / vale3.Open
```

```python
# Obtendo o valor percentual através do função .apply()

pct_c = (vale3.Close / vale3.Open).apply(lambda x:x-1)*100
display(pct_c.head(10))

# Variações positivas são > 0
# Variações negativas são < 0
```

```python
# Analisando melhor através de um gráfico de distribuição

# É útil para analisarmos se a VARIAÇÃO DO ATIVO está NORMALMENTE distribuída. Dependendo da resposta, nossas estratégias 
# deverão ser adaptadas.

# Cálculo utilizando escala logarítmica:

daily_log_returns = np.log(vale3.Close.pct_change()+1)

# np.log(): calcula o logaritmo natural de um número
# pct_change(): realiza o cálculo da variação percentural entre o elemento atual e o anterior de um dado conjunto de dados.
```

**Plotando em gráfico**

```python
import plotly
import plotly.offline as py
import plotly.graph_objs as go
import numpy as np
import plotly.graph_objs as go

py.init_notebook_mode(connected = True)

def simple_config_plot(fig, title):
    title = {'text':title, 'xanchor': 'center', 'yanchor': 'bottom', 'y':0,'x':0.5,}
    
    fig.update_layout(title=title,
          xaxis_rangeslider_visible = False, width = 854, height = 480,
          xaxis_showgrid = True, xaxis_gridwidth = 1, xaxis_gridcolor = '#E8E8E8',
          yaxis_showgrid = True, yaxis_gridwidth = 1, yaxis_gridcolor = '#E8E8E8',
          plot_bgcolor = 'rgba(0,0,0,0)')

def simple_plot(data, title):
    fig = go.Figure(data = data)
    
    simple_config_plot(fig, title)
    
    fig.show()

data = [go.Histogram(x = daily_log_returns)]

simple_plot(data, "1.1 VALE3.SA Histogram")
```

```python
# Para calcular a variaÃ§Ã£o percentual intradiÃ¡ria do ativo

# cÃ¡lculo: preÃ§o de fechamento / preÃ§o de abertura

# Com o pandas dividimos as colunas de forma direta.

vale3.Close / vale3.Open
```

```python
# Obtendo o valor percentual atravÃ©s do funÃ§Ã£o .apply()

pct_c = (vale3.Close / vale3.Open).apply(lambda x:x-1)*100
display(pct_c.head(10))

# VariaÃ§Ãµes positivas sÃ£o > 0
# VariaÃ§Ãµes negativas sÃ£o < 0
```

```python
# Analisando melhor atravÃ©s de um grÃ¡fico de distribuiÃ§Ã£o

# Ã‰ Ãºtil para analisarmos se a VARIAÃ‡ÃƒO DO ATIVO estÃ¡ NORMALMENTE distribuÃ­da. Dependendo da resposta, nossas estratÃ©gias 
# deverÃ£o ser adaptadas.

# CÃ¡lculo utilizando escala logarÃ­tmica:

daily_log_returns = np.log(vale3.Close.pct_change()+1)

# np.log(): calcula o logaritmo natural de um nÃºmero
# pct_change(): realiza o cÃ¡lculo da variaÃ§Ã£o percentural entre o elemento atual e o anterior de um dado conjunto de dados.
```

**Plotando em grÃ¡fico**

```python
import plotly
import plotly.offline as py
import plotly.graph_objs as go
import numpy as np
import plotly.graph_objs as go

py.init_notebook_mode(connected = True)

def simple_config_plot(fig, title):
    title = {'text':title, 'xanchor': 'center', 'yanchor': 'bottom', 'y':0,'x':0.5,}
    
    fig.update_layout(title=title,
          xaxis_rangeslider_visible = False, width = 854, height = 480,
          xaxis_showgrid = True, xaxis_gridwidth = 1, xaxis_gridcolor = '#E8E8E8',
          yaxis_showgrid = True, yaxis_gridwidth = 1, yaxis_gridcolor = '#E8E8E8',
          plot_bgcolor = 'rgba(0,0,0,0)')

def simple_plot(data, title):
    fig = go.Figure(data = data)
    
    simple_config_plot(fig, title)
    
    fig.show()

data = [go.Histogram(x = daily_log_returns)]

simple_plot(data, "1.1 VALE3.SA Histogram")
```
